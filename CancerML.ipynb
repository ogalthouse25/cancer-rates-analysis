```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame with columns 'State', 'Beds', and 'CostToChargeRatio'
# If your columns have different names, adjust accordingly

def create_state_dotplot(df):
    # Create the scatter plot
    plt.figure(figsize=(10, 6))
    # Changed to access the correct column names in hos DataFrame
    plt.scatter(df['Number of Beds'], df['Cost To Charge Ratio'], alpha=0.7)
    # Add labels and title
    plt.xlabel('Number of Beds')
    plt.ylabel('Cost to Charge Ratio')
    plt.title('State Healthcare: Beds vs Cost to Charge Ratio')

    # Add state labels to points (optional)
    # Changed to access the correct column names in hos DataFrame
    for i, state in enumerate(df['Provider State']):
        plt.annotate(state,
                    (df['Number of Beds'].iloc[i], df['Cost To Charge Ratio'].iloc[i]),
                    xytext=(5, 5),
                    textcoords='offset points',
                    fontsize=8)

    # Adjust layout and display
    plt.tight_layout()
    plt.grid(True, linestyle='--', alpha=0.7)
    return plt

# Example usage:
plot = create_state_dotplot(hos)
plot.savefig('state_dotplot.png')  # Save to file
plot.show()  # Display the plot
```
